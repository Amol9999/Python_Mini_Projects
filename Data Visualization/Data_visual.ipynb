##Variable Setup


```python
file_location = '/content/Crest_Factor_Outer_Race.csv'

image_name='Extracted_Featur_Crest_Factor_Outer_Race.png'

inc=51
```



##1.   **File Upload And Loading**



```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(file_location)
df.describe()
```



##2.   **X-axis**



```python
no_of_column=len(df.columns)
```

```python
x=[]
start=1
stop=51
for i in range(no_of_column):
  temp=list ( range( start , stop ) )
  x.append( temp )
  start=start+inc
  stop=stop+inc
```

##3.   **Plotting and labeling**


```python
#Setting the Size and resoution of 
plt.figure(figsize=(8, 5), dpi=300 )

# turning grid on
plt.grid()

# Creating scatter plot for each column in dataset corresponding to X values 
for column , x_index in zip( df.columns, x ):
  plt.scatter( list ( x_index ), df[column], label=str ( column ) , s=4 )

# Placing Title
plt.title('Comparsion between Healthy And Outer Race Fault')

# Axis labeling
plt.xlabel('Data Points')
plt.ylabel('Crest Factor Values')

# Legend
plt.legend()

# Saving the plot
plt.savefig(image_name , transparent=True )

# Plotting 
# plt.show()  
```
